```python
def parse_data(url):
    res = requests.get(url=url,headers=headers)
    res.encoding = 'utf-8'
    result = []
    result = res.json()['data']
    if result != None:
        client.douban.film.insert_many(result)
```

```python
for url in urls:
    parse_data(url)
```
